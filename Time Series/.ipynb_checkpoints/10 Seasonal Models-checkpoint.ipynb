# Seasonal ARIMA (SARIMA) and Extensions

In this lesson, we extend ARIMA models to account for seasonality using Seasonal ARIMA (SARIMA). SARIMA models capture seasonal patterns along with trend and noise.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

%matplotlib inline

# Generate synthetic seasonal time series data
np.random.seed(42)
dates = pd.date_range(start='2019-01-01', periods=730, freq='D')  # 2 years of daily data
trend = np.linspace(0, 20, 730)
seasonal = 10 * np.sin(2 * np.pi * dates.dayofyear / 365)
noise = np.random.normal(0, 1, 730)
data = trend + seasonal + noise

ts_df = pd.DataFrame({'Date': dates, 'Value': data}).set_index('Date').asfreq('D')

# Plot the seasonal time series
plt.figure(figsize=(12,6))
plt.plot(ts_df, label='Seasonal Time Series')
plt.title('Seasonal Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Fit a SARIMA model (example: SARIMA(1,1,1)x(1,1,1,12))
model = SARIMAX(ts_df['Value'], order=(1,1,1), seasonal_order=(1,1,1,12))
model_fit = model.fit(disp=False)
print(model_fit.summary())

# Forecast the next 30 days
forecast = model_fit.get_forecast(steps=30)
forecast_index = pd.date_range(start=ts_df.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')
forecast_series = pd.Series(forecast.predicted_mean, index=forecast_index)

plt.figure(figsize=(12,6))
plt.plot(ts_df, label='Original Data')
plt.plot(forecast_series, label='Forecast', marker='o')
plt.title('SARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## Conclusion

SARIMA models extend ARIMA by incorporating seasonal patterns, making them ideal for data with periodic fluctuations.
