# Time Series Visualization & Decomposition

In this lesson, we visualize a time series and decompose it into its components: trend, seasonality, and residuals.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

%matplotlib inline

# Create synthetic time series data with trend and seasonality
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=200, freq='D')
trend = np.linspace(0, 10, 200)
seasonality = 5 * np.sin(np.linspace(0, 4 * np.pi, 200))
noise = np.random.normal(0, 1, 200)
data = trend + seasonality + noise

ts_df = pd.DataFrame({'Date': dates, 'Value': data}).set_index('Date')

# Plot the time series
plt.figure(figsize=(12,6))
plt.plot(ts_df, label='Time Series')
plt.title('Time Series Visualization')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(ts_df, model='additive', period=30)

fig = decomposition.plot()
fig.set_size_inches(14, 10)
plt.show()
```

## Conclusion

We successfully visualized the time series and decomposed it to reveal the underlying trend, seasonality, and noise.
