# Autocorrelation and Partial Autocorrelation

In this lesson, we explore Autocorrelation (ACF) and Partial Autocorrelation (PACF) to identify lag relationships in time series data.

## Autocorrelation Function (ACF)

ACF measures the correlation between the time series and its lagged values.

## Partial Autocorrelation Function (PACF)

PACF measures the correlation between the series and its lagged values while controlling for shorter lags.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

%matplotlib inline

# Create synthetic time series data
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=150, freq='D')
data = np.sin(np.linspace(0, 3 * np.pi, 150)) + np.random.normal(0, 0.5, 150)
ts_df = pd.DataFrame({'Date': dates, 'Value': data}).set_index('Date')

# Plot the time series
plt.figure(figsize=(12,4))
plt.plot(ts_df, label='Time Series')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Plot ACF
fig, ax = plt.subplots(figsize=(12,4))
plot_acf(ts_df['Value'], ax=ax, lags=30)
plt.title('Autocorrelation Function (ACF)')
plt.show()

# Plot PACF
fig, ax = plt.subplots(figsize=(12,4))
plot_pacf(ts_df['Value'], ax=ax, lags=30, method='ywm')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

## Conclusion

ACF and PACF plots are essential tools for identifying the lag orders needed in AR and MA models.
