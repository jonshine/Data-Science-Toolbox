# Model Diagnostics and Forecasting

In this lesson, we cover how to evaluate a time series model by examining residuals and how to forecast future values.

## Residual Diagnostics

Model diagnostics help us determine whether the residuals behave like white noise.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA

%matplotlib inline

# Generate synthetic time series data
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=150, freq='D')
data = np.sin(np.linspace(0, 3 * np.pi, 150)) + np.random.normal(0, 0.5, 150)
ts_df = pd.DataFrame({'Date': dates, 'Value': data}).set_index('Date').asfreq('D')

# Fit an ARIMA(2,0,1) model
model = ARIMA(ts_df['Value'], order=(2, 0, 1))
model_fit = model.fit()

# Plot model diagnostics
model_fit.plot_diagnostics(figsize=(12,8))
plt.show()

# Forecast the next 10 steps
forecast = model_fit.forecast(steps=10)
plt.figure(figsize=(12,6))
plt.plot(ts_df, label='Original Data')
plt.plot(forecast.index, forecast, label='Forecast', marker='o')
plt.title('Forecasting with ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## Conclusion

By evaluating residuals and generating forecasts, we ensure that our model is appropriate and useful for predicting future values.
